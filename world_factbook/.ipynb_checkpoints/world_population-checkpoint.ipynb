```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline
```

```python
conn = sqlite3.connect(r'factbook.db') #create a sqlite3 Connection instance
q1= "SELECT * FROM sqlite_master WHERE type='table';" # create a query
table = pd.read_sql_query(q1, conn) #return information in the form of tables
print(table)
```

```python
# if we used the default sqlite3 querying method
cursor = conn.cursor()
results = cursor.execute(q1).fetchall()
```

```python
# Return minimum and maximum values of birth rate, death rate and migration rate
q2 = 'SELECT MIN(birth_rate), MAX(birth_rate), MIN(death_rate), MAX(death_rate),  MIN(migration_rate), MAX(migration_rate) FROM facts;'
pd.read_sql_query(q2, conn)
```

```python
#Let's take a look at the zero migration rate value.
q3 = 'SELECT * FROM facts WHERE migration_rate == (SELECT MIN(migration_rate) FROM facts)'
pd.read_sql_query(q3, conn)
```

#These are countries which are probably not attractive for newcomers or there is no statistical data

```python
#Let's take a look at the max migration rate value.
q4 = 'SELECT name FROM facts ORDER BY migration_rate LIMIT 10'
pd.read_sql_query(q4, conn)
```
